## 1. A general implementation of the echo state network with  explanation

Import modules and load the data set.

```python
# make sure plots are correctly displayed in the notebook
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as LA
import math
np.random.seed(seed=15)
```

```python
timeseries = np.loadtxt("2sin.txt")
training_data = timeseries
```

### Model Hyperparameters and Forcasting Horizon  
#### Forecasting Horizon
Before we can begin creating our model we have to decide the forecasting horizon of our model, that is, how many steps ahead do we want to predict? For the initial implementation we will predict 1 step ahead (k=1).  
#### Hyperparameters
For echo state networks we need to decide on the following values:  
1. **n** : This is the amount of data we want to use as our input. For example, if we want to predict timestep t+k and are at time t, then we can use the values for the function at time t, t-1,... t-n for create a larger input for u<sub>t</sub>. In theory, a larger input vector will allow for more accurate predictions  
2. **Nr** : This is the size of the resevoir (state) vector. A larger resevoir allows for more complex state representation. However, if the resovoir vector is too large it can make the model too complex, leading to overfitting
3. **spectral_radius_normalization** : This is the value that we regularize eigenvalues of Wr to. We choose 1 for this as having a spectral radius ~1 tends to give our model the *Echo State Property* which is needed for state convergence.  
4. **λ** : This valus is used in ridge regression to control model complexity. A higher value for this constant will lead to lower model complexity and help prevent overfitting.  
5. **N0** : The amount of data used by the model when optimizing the weights in ridge regression.

```python
k = 1 # forcasting horizon 
n = 0 # use t,t-1 ... t-n as Ut to predict t+k
Nr = 20 # res size
spectral_radius_normalization = 0.99 # regularization for Wr
λ = .02 # regularization for ridge regression
N0 = ((training_data.size - n) // k)-1 # amount of training data used in ridge regression
```

### Calculating W<sup>i</sup>  W<sup>r</sup> and normalizing the spectral radius of W<sup>r</sup>
W<sup>i</sup> is the matrix that transforms the input vector of length N<sub>i</sub> into a vector the same size as our reservoir (N<sub>r</sub>). To obtain this matrix we sample uniformly from the interval [-1,1]  
W<sup>r</sup> is linear transformation applied to the state vector x. As W<sup>r</sup> dictates the dynamics of the model we have tyo be more careful when we initialize it. We start by sampling from [-1,1] but then we regularize the spectral radius of W<sup>r</sup> to be the spectral_radius constant (hyperparameter), which is ~1.  
This can be done with the calculation:  
W<sup>r</sup> = spectral_radius_normalization * (W<sup>r</sup> / spectral_radius)  
As the spectral_radius_normalization variable is ~1 W<sup>r</sup> has a spectral radius of ~1 after normalization.  
This step is important to increase the likelyhood that the model has the *Echo State Property*.

```python
Wi = np.random.uniform(-1,1,(Nr,n+1))
Wr = np.random.uniform(-1,1,(Nr,Nr))

eigenvals = LA.eigvals(Wr)
spectral_radius = np.abs(eigenvals).max()
Wr = Wr * (1/spectral_radius) * spectral_radius_normalization
```

### Collecting states in X and targets in t
The next step in our model is obtaining X and t (which we need for ridge regression).  
To do this we:  
1. Set the initial state of x to be the 0 vector.
2. Feed our echo state network with our input data and collect the resulting states in X.
3. Collect the target value associated with each state in t.  

The state is updated using the function:  

x<sub>t</sub> = tanh( **W**<sup>r</sup>x<sub>t-1</sub> + **W**<sup>i</sup>u<sub>t</sub>)  

This is run for each input, and the resulting states are collected in **X** and corresponding targets are collected in **t**.
This results in **X** being a N0 by N<sup>r</sup> matrix and **t** being a vector of length N<sup>r</sup>.

```python
X = [] # data matrix
t = np.array([]) #target vector
index = n
x = np.zeros([Nr,1]) # resevoir, initialized to the zero vector

for i in range(0,N0): #keep going as long as there are points in the training data that still have targets
    u = np.array(training_data[index-n:index+1]).reshape(-1,1)
    x = np.tanh(np.matmul(Wr, x) + np.matmul(Wi, u)) # state update function
    t = np.append(t,training_data[index+k])
    X.append(x)
    index += k
    
X = np.array(X).reshape(-1,Nr)
t = t.reshape(-1,1)
```

### Ridge regression using **X** and **t**
We now have to make a ridge regression model that maps our matrix of states **X** to our targets **t**.  
We need to solve the least-square problem for W<sup>o</sup>.
That is, we need to find the value of matrix **W** such that our loss function:  

1/2||**XW** - **t**||<sup>2</sup> + λ/2 ||**W**||<sup>2</sup>  

is minimized.
This can be done by solving:  
(**X**<sup>T</sup>**X** + λ**I**)<sup>-1</sup> + λ/2||**W**||<sup>2</sup>  
where I is the identity matrix, then:  
**W**<sup>o</sup> = **W**<sup>T</sup>  
Then we can use **W**<sup>o</sup> map states into outputs:  
y<sub>t</sub> = **W**<sup>o</sup> **x**<sub>t-1</sub>

```python
X = np.insert(X, X.shape[1], np.ones(X.shape[0]), axis=1) # add bias to X vector
XtransposeX = np.matmul(X.transpose(), X) # (X^T)X
regularization_matrix = λ * np.identity(Nr+1) # λI 
inverted_matrix = LA.inv(XtransposeX + regularization_matrix) # ((X^T)X + λI)^-1
W = np.matmul( inverted_matrix , np.matmul(X.transpose(),t)) # ((X^T)X + λI)^-1 (X^T) t
W0 = W.transpose() # W0 = W^T
```

## 2. Metrics and visualization for evaluating the performance of the model
Across all sections of the report the model will be evaulated in the following ways:

    2. Using metrics like mean-squared error (MSE) and and Normalized Root MSE (NRMSE)

### Prediction Function
To evaulate our model we need it to make predictions on a data set (that we can compare to targets).  
This is done by taking input data, running the input through our state update function and then mapping the resulting states to outputs using our ridge regression model. This is done by multiplying our states with the matrix that represents our ridge regresssion **W**<sup>o</sup>.  
To make predictions on a data set:  
1. The state is initialized to the zero vector.  
2. The inputs (and states) are run through the state update equation.  
3. The states are mapped to oututs through matrix vector multiplcation with W0
The function also returns the actual targets (**t**) so that it is easier to calculate the error between predictions and targets.

```python
def make_prediction(data,n,Nr,k,W0,Wi,Wr):
    index = n
    x = np.zeros([Nr,1])
    predictions = np.array([])
    indices = np.array([])
    targets = np.array([])
    
    while index+k < data.size:
        
        u = np.array(data[index-n:index+1]).reshape(-1,1)
        x = np.tanh(np.matmul(Wr, x) + np.matmul(Wi, u))
        
        predictions = np.append(predictions, np.matmul(W0,np.append(x,1)))
        indices = np.append(indices, index+k)
        targets = np.append(targets, data[index+k])
        index+=k
    return (predictions,indices,targets)
pred,indices,t = make_prediction(timeseries,n,Nr,k,W0,Wi,Wr)
```

### Graphing  
We can graph our prediction (red) against the dataset (blue).
The fit seems to be pretty good, but we can notice that our model slightly overpredicts the tips of each wave. Note that this model has not been cross-validated or tested aganist a seperate testing data partition, that is done in later sections in the notebook. 

```python
x_cords = np.arange(0, timeseries.size , 1, dtype=int)
plt.plot(indices[100:1000], pred[100:1000], color="r")
plt.plot(indices[100:1000],t[100:1000], color="b")
plt.legend( ["Prediction" , "Target"])
plt.show()
```

### MSE and NRMSE
MSE = < || **t** - **z** ||<sup>2</sup> >  = 1/N Σ(**t**-**z**)<sup>2</sup> where N is the size of t and z

NRMSE = sqrt(MSE) / var(**t**)    

NRMSE can be used to compare fits across datasets (and forecasting horizons) as it factors in the variance of the data.

```python
def calculate_eror(pred,t):
    z = np.subtract(pred,t)
    MSE = np.sum(np.square(z))/pred.size
    var_t = np.var(t.reshape(-1))
    NRMSE = np.sqrt(MSE)/var_t
    return MSE,NRMSE

MSE,NRMSE = calculate_eror(pred,t)

print("Mean-squared error is: ")
print(MSE)
print("\n")
print("Normalized Root MSE is: ")
print(NRMSE)
```

## 3. The ESN implemented as a single function
In this section we implement the echo state network as a function so that it can be reused for the hyperparameter search experiments in later sections. 

This is the same implementation described in section 1, compiled into a single function. The only difference is that going forward the hyperparameter N0 is now comtrolled by the size of the training data set passed to the ESN.

```python
def ESN(training_data,k,n,Nr,spectral_radius_normalization,λ):
    
    ## Initialize and normalize the values of weight matrices
    Wi = np.random.uniform(-1,1,(Nr,n+1))
    Wr = np.random.uniform(-1,1,(Nr,Nr))
    eigenvals = LA.eigvals(Wr)
    spectral_radius = np.abs(eigenvals).max()
    Wr = Wr * (1/spectral_radius) * spectral_radius_normalization
    
    ## Calculate data matrix X and target vecotor t
    X = [] # data matrix
    t = np.array([]) #target vector
    index = n
    x = np.zeros([Nr,1]) # resevoir, initialized to the zero vector
    
    while index < training_data.size - k : #keep going as long as there are points in the training data that still have targets
        u = np.array(training_data[index-n:index+1]).reshape(-1,1)
        x = np.tanh(np.matmul(Wr, x) + np.matmul(Wi, u)) # state update function
        t = np.append(t,training_data[index+k])
        X.append(x)
        index += k
    
    X = np.array(X).reshape(-1,Nr)
    t = t.reshape(-1,1)
    
    ## calculate W0 via ridge regression
    X = np.insert(X, X.shape[1], np.ones(X.shape[0]), axis=1) # add bias to X vector
    XtransposeX = np.matmul(X.transpose(), X) # (X^T)X
    regularization_matrix = λ * np.identity(Nr+1) # λI 
    inverted_matrix = LA.inv(XtransposeX + regularization_matrix) # ((X^T)X + λI)^-1
    W = np.matmul( inverted_matrix , np.matmul(X.transpose(),t)) # ((X^T)X + λI)^-1 (X^T) t
    W0 = W.transpose() # W0 = W^T
    return (W0,Wr,Wi,t)
```

## 4. Cross-validation on the lorenz.txt data set with different values of k

To learn the hyperparameters of our model we use *cross-validation*. That is, we split the data sets into three partitions:  
*training data*, *validation data* and *testing data*. In this section the hyperparameters we explore are the amount of data used for traning and the resevoir size.  
**Training data** is used for the initial training of the model (populating data matrix **X** with states, and calculating W<sup>O</sup>).  
**Validation data** is used for finding which hyperparameters minimize error in the model. In this case, we use grid search to search the hyperparameter space. Training and Validation data sets should be seperate paritions because we don't want to find the hyperparameters that optimize performance on the training data, we want the model to *generalize* (perform well on data outside its training set).  
**Testing data** is what we use to evaluate how well the model *generalizes*. As this is data the model was not trained on, or cross-validated on, it should provide insight on on well the model will perform on data it has never seen before.  
The **split_data** function is used to divide the data sets into these three partitions. 

```python
timeseries = np.loadtxt("lorenz.txt")

def split_data(data,training,validation,testing):
    
    diff = 1 - training - validation - testing
    training_points = math.floor(data.size*training)
    diff = math.floor(data.size*diff) + training_points
    validation_points = math.floor(data.size*validation) + diff
    testing_points = math.floor(data.size*testing) + validation_points
    
    return (data[0:training_points], data[training_points:validation_points],data[validation_points: data.size])
```

### Hyperparameter grid search
To find the lowest error combination we search an n-dimensional table of hyperparameter values (where n is the number of hyperparameters we are searching). Our grid search runs each configuration w times and averages the error across each run. This is to reduce the effect of the randomization of W<sup>i</sup> and W<sup>r</sup> on the outputed error for each configuration in the grid search, which should give us a more accurate measure of how the values of the hyperparameters we are tuning effect the error of the model.

```python
k=10

def search_hyperparams():
    data_amount = (.2,.3,.4,.5,.6)
    res_sizes = (20,30,40,50,60,70)
    results = []
    for amt in data_amount:
        training_data,validation_data,testing_data = split_data(timeseries, amt,.2,.2)
        for res_size in res_sizes:
            mean_error = 0
            normal_mean_error = 0
            for w in range(0,3):
                W0,Wr,Wi,t =  ESN(training_data, k,n,res_size,spectral_radius_normalization,λ)
                pred,indices,t = make_prediction(validation_data,n,res_size,k,W0,Wi,Wr)
                mean_square, normal_mean_square = calculate_eror(pred,t)
                mean_error+=mean_square
                normal_mean_error+=normal_mean_square
            results.append((amt,res_size,mean_error/3.0,normal_mean_error/3.0))
    results.sort( key=lambda tup: tup[2])
    return results,results[0]
results,best_config = search_hyperparams()
```

### Evaluating the effects of hyperparameters
The surface plot shows a clear trend that large values of k lead to more data being needed to lower error during cross validation. Even with large partitions of training data, the model struggles to minimize error with large values of k. This makes sense as there are not that many points available for training with large values for k (as the model steps through the dataset in larger increments). All models show preference for larger resevoir sizes, with most being in the range of 50 to 70.  
All surface plots have **hyperparameters as the z and y axes**, and **NRMSE is the z axis**.

```python
def displayHP(results):
    results_array = np.array(results)
    fig = plt.figure(constrained_layout=True) 
    ax = fig.add_subplot(111, projection='3d') 
    ax.plot_trisurf(results_array[:,0], results_array[:,1], results_array[:,3], color='white', edgecolors='grey', alpha=0.5) 
    ax.scatter(results_array[:,0], results_array[:,1], results_array[:,3], c='red')
    plt.ylabel("Resevoir size")
    plt.xlabel("Proportion of data used")
    plt.show()
displayHP(results)

print("The best configuration had a resevoir size of: ", best_config[1])
print("The best used ",best_config[0]*100.0, " percent of the data set for training")
print("The best configuration had a NRMSE of: ", best_config[3])
```

## 6. Learning the regularization coefficent for lorenz.txt
Now that we have a better idea of the correct configuration of the model with regards to training data and resevoir size, we can move on to tuning the regularization in the model. High values of λ represent large amount of regularization in the model. This will help us create a good fit, improve the generalization of the model and prevent the model from becoming overly complex. In our ESN the regularization is in the ridge regression step. That is, we are adding regularization to our linear regression that maps states to outputs (W<sup>O</sup> ).
We acomplish this by cross validation of on the regularization coefficent λ.
This is demonstrated below on the lorenz data set, with the ESN set to predict k=10 steps ahead.

```python
### configuring hyper params discovered in the previous section
timeseries = np.loadtxt("lorenz.txt")

k=10
training_data,validation_data,testing_data = split_data(timeseries, .4,.2,.2)
optimized_res_size = 60
```

```python
def search_regularization_param(reg_range=np.arange(0.001, 20.0, .1)):
    results = []
    reg_opt = reg_range
    
    for reg in reg_opt:
        W0,Wr,Wi,t =  ESN(training_data, k,n,optimized_res_size,spectral_radius_normalization,reg)
        pred,indices,t = make_prediction(validation_data,n,optimized_res_size,k,W0,Wi,Wr)
        mean_square, normal_mean_square = calculate_eror(pred,t)
        results.append((reg ,mean_square,normal_mean_square))
    return results
```

#### Visualizing how regularization effects model error
We can graph how changing λ changes the error in our model. It seems that on the lorenz.txt data set there is a small preference for lower values of λ, but also that small changes in the value of λ lead to big changes in model error.

```python
results = search_regularization_param()
arr = np.array(results)
plt.plot(arr[:,0], arr[:,2], color="r")
plt.xlabel("Regularization constant")
plt.ylabel("NRMSE")
plt.show()

results.sort( key=lambda tup: tup[2])

print("The best regularization constant was: ")
print(results[0][0])
```

## 8. Exhaustive hyperparameter search
While sections 4-7 provide good insight into the theory of grid search and cross-validation, to effectively tune hyperparameters you need to tune hyperparmeters together (and across large ranges of values of each hyperparameter).  
This is **computationally intensive** so I have only included an example with one value of k over a restricted range (for the hyperparameters) in the notebook, but I have provided results from these experiments in CSV files under the folder **hyperparameter search data**.  
Optimizing all hyperparameters together leads to different (and better) combinations of hyperparameters than individual hyperparameter optimization, but is much more computationally intensive.

```python
### configuring hyper params discovered in the previous section
timeseries = np.loadtxt("2sin.txt")
k=3
def exhaustive_search_hyperparams():
    data_amount = (.2,.3,.4,.5,.6)
    res_sizes = np.arange(1,100,10, dtype=int)
    reg_opt = np.arange(.001,1.5,.1)
    results = []
    for reg in reg_opt:
        for amt in data_amount:
            training_data,validation_data,testing_data = split_data(timeseries, amt,.2,.2)
            for res_size in res_sizes:
                mean_error = 0
                normal_mean_error = 0
                for w in range(0,3):
                    W0,Wr,Wi,t =  ESN(training_data, k,n,res_size,spectral_radius_normalization,reg)
                    pred,indices,t = make_prediction(validation_data,n,res_size,k,W0,Wi,Wr)
                    mean_square, normal_mean_square = calculate_eror(pred,t)
                    mean_error+=mean_square
                    normal_mean_error+=normal_mean_square
                results.append((amt,res_size,reg, mean_error/3.0,normal_mean_error/3.0))
    results.sort( key=lambda tup: tup[4])
    return results,results[0]
results,best_config = exhaustive_search_hyperparams()
print("The best configuration had a resevoir size of " , best_config[1], ", a regularization constant of " , best_config[2], "\n and used ", best_config[0]*100, " percent of the data.")
```

## 9. Evaluating the performance and generalization of the models with different data and forecasting horizons
In this section we compare the performance of our model across data sets and forecasting horizons. In this section we will use the last partition of our data (testing) to see if our trained and cross-validated models generalize effectively. We examine these models by graphing our predictions aganist the data, and by comparing error metrics like MSE and NRMSE.  
We use NRMSE to compare error as it can be used to compare error across models with different prediction horizons and data sets (as NRMSE is normalized for the varaiance in each data set to allow for comparisons across data sets).

### Error at the start of the testing data   
In many of our models, most of the error is in the first few predictions. I suspect this is because the state vector is initialized to zero, and needs to converge to a useful value before it results in good predictions. Therefore, when evaluating the models I created a new function **calculate_eror_without_start** which will calculate the error excluding the first few predictions. Many model's NRMSE is significantly reduced by excluding the error on the models first few predictions.  
This is especially true with models trained on 2sin.txt because the testing data set is smaller. Therefore, the error on the first few points before state vector converges represents a larger proportion of the error in the NRMSE value.  
By excluding the first few points from the error caluclation we can gain a better understanding of the error in the model **in the long run**.  
**Both metrics (with the frist few prediction included and excluded from error) are given for each model (where appropriate).**

```python
def calculate_eror_without_start(pred,t,drop):
    z = np.subtract(pred,t)
    z = z[drop-1:z.shape[0]]
    MSE = np.sum(np.square(z))/z.size
    var_t = np.var(t.reshape(-1))
    NRMSE = np.sqrt(MSE)/var_t
    
    return MSE,NRMSE
```

#### Optimized model for **2sin.txt with a forecasting horizon of 3.**

```python
timeseries = np.loadtxt("2sin.txt")
training_data,validation_data,testing_data = split_data(timeseries, .5,.2,.2)
W0,Wr,Wi,t =  ESN(training_data, 3,0,81,spectral_radius_normalization,.001)
pred,indices,t = make_prediction(testing_data,0,81,3,W0,Wi,Wr)
mean_square, normal_mean_square = calculate_eror(pred,t)
plt.plot(indices, pred, color="r")
plt.plot(indices, t, color="b")
plt.legend( ["Prediction" , "Target"])
plt.show()
```

```python
print("MSE: ")
print(mean_square)
print("NRMSE: ")
print(normal_mean_square)

print("\nError after the state vector converges: \n")

mean_square, normal_mean_square = calculate_eror_without_start(pred,t,10)

print("MSE: ")
print(mean_square)
print("NRMSE: ")
print(normal_mean_square)
```

Intuitively, I would expect larger prediction horizons to have more error because it should be harder to predict distantly into the future. Also many of our models have high error while the state vector converges from its initalized value to a useful value in the first few predictions. This makes sense because the initial value fo the state vector is not helpful for the ridge regression model to produce outputs. Only once the state vector converges to a useful state will the ridge regression model be able to output accurate predictions.

## 10. Investigating using larger inputs for u<sub>t</sub>
In this section we do cross-validation hyperparameter grid search with the size of the input u<sub>t</sub> as an additional hyperparameter. We now use t,t-1,... t-n as the input vector for the state update. 

```python
timeseries = np.loadtxt("lorenz.txt")
k=10
def input_size_hyperparams():
    input_sizes = np.arange(0,10,1, dtype=int)
    training_data,validation_data,testing_data = split_data(timeseries, .5,.2,.2)
    res_sizes = np.arange(1,100,10, dtype=int)
    reg_opt = np.arange(.001,.5,.1)
    results = []
    for reg in reg_opt:
        for input_size in input_sizes:
            for res_size in res_sizes:
                mean_error = 0
                normal_mean_error = 0
                for w in range(0,3):
                    W0,Wr,Wi,t =  ESN(training_data, k,input_size,res_size,spectral_radius_normalization,reg)
                    pred,indices,t = make_prediction(validation_data,input_size,res_size,k,W0,Wi,Wr)
                    mean_square, normal_mean_square = calculate_eror(pred,t)
                    mean_error+=mean_square
                    normal_mean_error+=normal_mean_square
                results.append((input_size,res_size,reg, mean_error/3.0,normal_mean_error/3.0))
    results.sort( key=lambda tup: tup[4])
    return results,results[0]
results,best_config = input_size_hyperparams()
```

```python
print("The best configuration was")
print("input vector size was: " )
print(best_config[0]+1)
print("resevoir size: " )
print(best_config[1])
print("regularization constant: " )
print(best_config[2])
print("NRMSE on the best fit was")
print(best_config[4])
```

This cross-validation experiment evidences that a larger input vector can lead to lower error models. Intuitively, it makes sense that a larger input vector would be helpful for the model, as it provides much more information as to whether the function is increasing or decreasing.  
For example, [.45,.47,.5] gives you a much stronger indication that the function is increasing than [.5].

## 11. Investigating using different values for the spectral radius

How does the spectral radius effect the model? It is quickly apparent that large spectral radiuses (pictured below) lead to poor performance and the network not being able to properly learn the function that underlies the data.  
Cross-validation can be used to see how changing the spectral radius effects the performance of the model. 

#### Large spectral radius leads to poor performance

```python
spectral_radius_normalization = 20.0
k=10
n=0
res_size = 80
λ = .03
training_data,validation_data,testing_data = split_data(timeseries, .5,.2,.2)

W0,Wr,Wi,t =  ESN(training_data, k,n,res_size,spectral_radius_normalization,λ)
pred,indices,t = make_prediction(testing_data,n,res_size,k,W0,Wi,Wr)
mean_square, normal_mean_square = calculate_eror(pred,t)

plt.plot(indices[0:100], pred[0:100], color="r")
plt.plot(indices[0:100], t[0:100], color="b")
plt.legend( ["Prediction" , "Target"])
plt.show()
```

Clearly, this model is not effective, especially compared to our models in section 9.

#### Cross-validation on the spectral radius

```python
def cross_validate_spectral_radius(spectral_radius_options):
    results = []
    
    for spectral_radius in spectral_radius_options:
        MSE_sum = 0;
        NRMSE_sum = 0;
        for i in range(0,5):
            W0,Wr,Wi,t =  ESN(training_data, k,n,res_size,spectral_radius,λ)
            pred,indices,t = make_prediction(validation_data,n,res_size,k,W0,Wi,Wr)
            mean_square, normal_mean_square = calculate_eror(pred,t)
            MSE_sum = MSE_sum + mean_square
            NRMSE_sum = NRMSE_sum + normal_mean_square
        results.append((spectral_radius,MSE_sum/5.0,NRMSE_sum/5.0))
    return results

#sweep across the range .6 to 1.2
results = cross_validate_spectral_radius(np.arange(-3.0,3,.1))
results_array = np.array(results)
plt.plot(results_array[:,0], results_array[:,2], color="r")
plt.xlabel("Spectral Radius Normalization Constant")
plt.ylabel("NRMSE")
plt.show()

results.sort( key=lambda tup: tup[2])
spectral_radius_normalization = 0.99 # reset to allow  for earlier cells to be rerun
```

#### Analyzing the effects of the spectral radius  
The first thing to notice is that the error has symmetry about 0. This makes sense algebraically as for normalizing the spectral radius only the absolute value of the spectral radius normalization constant should matter. As for any matrix **X**, ρ(**X**) = ρ((-1) **X**).  
It is also clear that values for the spectral radius close to 0 also result in high error. The optimal range for the spectral radius seems to be [1,.5] (or [-.5, -1]). This aligns with our expectation that it is very likely that the model will have the Echo State Property when the spectral radius of W<sup>r</sup> is around (or a little less than) 1.

Considering that  W<sup>r</sup> and W<sup>i</sup> are both randomized (within a range and W<sup>r</sup> is normalized to control the dynamics of the model) the ability of the ridge regression model to accurately map the resulting states to outputs is very impressive.
