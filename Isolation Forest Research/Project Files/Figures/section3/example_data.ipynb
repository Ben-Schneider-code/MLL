```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cov1 = np.matrix([[1,0],[0,1]])
cov2 = np.matrix([[2,0],[0,2]])

mean1 = np.array([0,0])
mean2 = np.array([-7,-7])
```

```python
dist1 = np.random.multivariate_normal(mean1, cov1, size=400)
dist2 = np.random.multivariate_normal(mean2, cov2, size=800)
```

```python
dist1.shape
```

```python
D = np.append(dist1,dist2, axis=0)
```

```python

plt.scatter(dist1[:, 0], dist1[:, 1], c= "black")
plt.scatter(dist2[:, 0], dist2[:, 1], c ="red")
```

```python
counts, bins = np.histogram(D[:, 0])
plt.stairs(counts, bins)
```
